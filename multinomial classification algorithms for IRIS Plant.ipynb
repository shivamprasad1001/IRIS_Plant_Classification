```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
dataSet = pd.read_csv(r"iris.csv")
```

```python
dataSet.head()
# dataSet.info()
```

```python
dataSet["output"].unique()
```

```python
sns.pairplot(data=dataSet, hue="output")
```

```python
x = dataSet.iloc[:,:-1]
y = dataSet["output"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
SoftMax_model = LogisticRegression(multi_class="multinomial")
```

```python
SoftMax_model.fit(x_train, y_train)
```

```python
SoftMax_model.score(x_test,y_test)*100
```

```python
import numpy as np

# Ensure you have the correct feature names, assuming they are ['sepal_len', 'sepal_width', 'petal_len', 'petal_width']
feature_names = ['sepal_len', 'sepal_width', 'petal_len', 'petal_width']

# Create new data with the correct shape and feature names
new_data = np.array([[0.4, 1.4, 0.5, 1.5]])
new_data_df = pd.DataFrame(new_data, columns=feature_names)

# Predict using the SoftMax_model
prediction = SoftMax_model.predict(new_data_df)[0]
prediction

```
